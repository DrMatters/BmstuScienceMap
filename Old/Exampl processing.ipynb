```python
authors = begin_table['Авторы']
```

```python
def process_author_string(author_string):
    splitted = author_string.split('\n')
    str_list = []
    for substr in splitted:
        temp = ' '.join(substr.split())
        str_list.append(temp)
    str_list = list(filter(None, str_list))
    return str_list
    
```
